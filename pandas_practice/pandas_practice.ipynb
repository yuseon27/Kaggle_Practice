```python
type(df[df['키'] > 180])
```

#### 1. 맨 뒤에 출력할 column 붙이기

```python
df[df['키'] > 180]['이름']
```

```python
df[df['키'] > 180][['이름','성별']]
```

```python
df.loc[df['키'] > 180, '이름':'성별']
```

```python
df.loc[df['키'] > 180, ['이름','성별']]
```
