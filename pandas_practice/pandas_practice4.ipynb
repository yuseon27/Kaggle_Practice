```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://bit.ly/ds-korean-idol')
```

### 4. 피벗 테이블(Pivot Table)

```python
df.head()
```

```python
# index는 행 인덱스
# column은 열 인덱스
# values는 조회하고 싶은 값
pd.pivot_table(df, index='소속사', columns='혈액형', values='키')
```

```python
pd.pivot_table(df, index='소속사', columns='혈액형', values='키', aggfunc=np.mean)
```

```python
pd.pivot_table(df, index='소속사', columns='혈액형', values='키', aggfunc=np.sum)
```

### 5. GroupBy (그룹으로 묶어 보기)

산술 통계는 가능한 열만 출력됨 - sum, mean, std, ...

```python
df.groupby('소속사')
```

```python
df.groupby('소속사').count()
```

```python
df.groupby('소속사').mean()
```

```python
df.groupby('소속사').sum()
```

```python
df.groupby('소속사').mean()
```

```python
df.groupby('소속사')['키'].mean()
```

### 6. Multi-Index (복합 인덱스)

#### 6-1. Multi-Index 적용

행 인덱스를 복합적으로 구성하고 싶은 경우 인덱스를 리스트로 만들어 줍니다.

```python
df.groupby(['혈액형', '성별']).mean()
```

#### 6-2. Multi-Index 데이터 프레임을 피벗 테이블로 변환

```python
df2 = df.groupby(['혈액형', '성별']).mean()
```

```python
df2
```

```python
df2.unstack('혈액형')
```

```python
df2.unstack('성별')
```

#### 6-3. 인덱스 초기화 (reset_index)

reset_index()는 Multi-Index로 구성된 데이터 프레임의 **인덱스를 초기화**해 줍니다.

```python
df2 = df2.reset_index()
```
